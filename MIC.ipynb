```python
#手続き「平均相互情報量」
def ave_MIC(Pp,tp,fn,fp,tn):
    
    #正規化 #確率
    Stp=tp*Pp
    Sfn=fn*Pp
    Sfp=fp*(1-Pp)
    Stn=tn*(1-Pp)
    
    Pa=Stp+Sfn
    Pb=Stp+Sfp
    
    #条件付き確率
    P_a_given_b=Stp/(Stp+Sfp)
    P_a_given_not_b=Sfn/(Sfn+Stn)
    
    #相互情報量
    MIC_a_given_b=math.log2(P_a_given_b/Pa)
    MIC_a_given_not_b=math.log2(P_a_given_not_b/Pa)
    MIC_not_a_given_b=math.log2((1-P_a_given_b)/(1-Pa))
    MIC_not_a_given_not_b=math.log2((1-P_a_given_not_b)/(1-Pa))
    print("相互情報量: ", MIC_a_given_b,MIC_a_given_not_b,MIC_not_a_given_b,MIC_not_a_given_not_b)
                                   
    #同時確率
    P_a_and_b=Stp
    P_a_and_not_b=Sfn
    P_not_a_and_b=Sfp
    P_not_a_and_not_b=Stn
    print("同時確率: ", P_a_and_b,P_a_and_not_b,P_not_a_and_b,P_not_a_and_not_b)
    
    #同時確率*情報量
    P_times_MIC_1=P_a_and_b*MIC_a_given_b
    P_times_MIC_2=P_a_and_not_b*MIC_a_given_not_b
    P_times_MIC_3=P_not_a_and_b*MIC_not_a_given_b
    P_times_MIC_4=P_not_a_and_not_b*MIC_not_a_given_not_b
    print("同時確率*情報量: ", P_times_MIC_1,P_times_MIC_2,P_times_MIC_3,P_times_MIC_4)
    
    #平均相互情報量
    ave_MIC = P_times_MIC_1 + P_times_MIC_2 + P_times_MIC_3 + P_times_MIC_4
    print("平均相互情報量: ", ave_MIC)
```

```python
#ave_MIC手続き確認
N=10000
Pp=1/N #真に陽性である確率
tp=0.999
fn=0.001
fp=0.01
tn=0.99
ave_MIC(Pp,tp,fn,fp,tn)
```

```python
#手続き「F1値」
def cal_F1(Pp,tp,fn,fp,tn):
    
    #正規化 #確率
    Stp=tp*Pp
    Sfn=fn*Pp
    Sfp=fp*(1-Pp)
    Stn=tn*(1-Pp)
    
    precision=Stp/(Stp+Sfp)
    recall=Stp/(Stp+Sfn)
    print("P:", precision, "R:", recall)
    F1=2*precision*recall/(precision+recall)
    print("F1:", F1)
```

```python
#cal_F1手続き確認
N=10000
Pp=1/N
tp=0.25
fn=0.75
fp=0.5
tn=0.5
cal_F1(Pp,tp,fn,fp,tn)
```
